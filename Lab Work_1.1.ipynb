Ans. Structured Data
- Data that is organized in rows and columns (like in tables or spreadsheets).
- Easily stored in relational databases (SQL).
- Examples:
  - Customer databases
  - Sales transactions
  - Employee records
  - Sensor data

    Unstructured Data
- Data that doesn't follow a predefined format or schema.
- Typically text-heavy, image-based, or audio/video content.
- Examples:
  - Emails
  - Social media posts
  - Images and videos
  - Medical imaging and clinical notes

    Applications of Statistics in Various Domains:

Healthcare
- Predict disease risk using patient history (logistic regression).
- Clinical trial analysis using hypothesis testing.
- Resource optimization in hospitals (queueing models, survival analysis).

Business
- Forecasting sales and revenue using time series models.
- Risk management and credit scoring using regression or classification.
- Inventory optimization using probability distributions.

Marketing
- Customer segmentation (clustering).
- A/B testing for ad performance.
- Churn prediction models using statistical inference.

```python
# Q.3
# - Create a Pandas DataFrame with the following columns:
# - Name (categorical)
# - Age (numerical)
# - Department (categorical)
# - Salary (numerical)

import pandas as pd
import numpy as np
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 40, 45],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'Sales'],
    'Salary': [50000, 60000, 70000, 80000, 90000]
})
print(df)
print("-------------------------------------")
# - Use types to show data types of each column.

print(df.dtypes)
print("-------------------------------------")
# - Use select _dtypes) to filter numerical and categorical data.
numerical_data = df.select_dtypes(include=[np.number])
print(f"Numerical Data: {numerical_data}")

categorical_data = df.select_dtypes(include=[object])
print(f"Categorical Data: {categorical_data}")
```

```python
# Q.4
# - Use Matplotlib or Seaborn to plot:
# - Histogram for age distribution.
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(df['Age'], bins = 10, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# - Bar chart for department count.
plt.bar(df['Department'], df['Salary'], color='skyblue')
plt.title('Department Salary Count')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.show()

# - Pie chart of department percentages.
plt.pie(df['Salary'], labels=df['Department'], autopct='%.1f%%', startangle=140)
plt.title('Department Salary Distribution')
plt.show()

# - Boxplot for salary.
plt.boxplot(df['Salary'], vert=True)
plt.title('Salary Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.show()

# - Write 1-2 lines describing the pattern seen in each chart.
```

```python
# Q.5
# - Generate a dataset of 1,000 random numbers (e.g., test scores) using NumPy.

random_data = np.random.randint(0, 100, size=1000)
print("Random Data Sample:", random_data[:10])

# - Take a random sample of 100 entries.

sample_data = np.random.choice(random_data, size=1000, replace=False)
print("Sample Data Sample:", sample_data[:10])

# - Calculate and compare the mean and standard deviation of:
# - Entire population

population_mean = np.mean(random_data)
population_std = np.std(random_data)
print(f"Population Mean: {population_mean}, Population Std Dev: {population_std}")

# - Sample

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data)
print(f"Sample Mean: {sample_mean}, Sample Std Dev: {sample_std}")

# - Explain the difference between population and sample statistics.
# POPULATION:-
# Population statistics refer to the entire dataset, while sample statistics are derived from a subset of the population. 
# The sample mean and standard deviation may differ from the population values due to sampling variability, but they 
# provide estimates of the population parameters.

# SAMPLE:-
# Sample statistics are used to infer characteristics of the population based on a smaller subset.
```
