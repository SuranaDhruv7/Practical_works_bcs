```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Q.1
# - Generate a normally distributed dataset using numpy.random.normal().

data = np.random.normal(loc=50, scale=20, size=100)
print(data[:10])

# - Plot a histogram and KDE using seaborn.histplot() or sns.kdeplot().
# - Overlay the mean and standard deviation lines on the plot.

mean = np.mean(data)
std_dev = np.std(data)
plt.axvline(mean, color='red', label='Mean')

colors = ['green', 'blue', 'orange']
for i in range(1,4):
    plt.axvline(mean + i*std_dev, color=colors[i-1], label='Mean + 1 Std Dev')
    plt.axvline(mean - i*std_dev, color=colors[i-1], label='Mean - 1 Std Dev')
# change the color only of kdeplot to any color you like
sns.histplot(data, kde=True, bins=30)
plt.title('Histogram with KDE')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# - Use scipy.stats.norm.pdf() to show the bell curve (optional).

from scipy.stats import norm
pdf = norm.pdf(data,loc=data.mean(),scale=data.std())
plt.plot(data, pdf, color='green', label='Normal PDF')
plt.title("Histogram + PDF using scipy.stats.norm.pdf()")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
# Q.2
# - Load a sample dataset and select a numeric column.

df = sns.load_dataset("tips")
print(df)

numeric_column = df['total_bill']
print(numeric_column)

# - Calculate using numpy.percentile() or df.quantile():
# - 25th percentile
# - 50th percentile (median)
# - 75th percentile

q25 = np.percentile(numeric_column, 25)
print(q25)

q50 = np.percentile(numeric_column, 50)
print(q50)

q75 = np.percentile(numeric_column, 75)
print(q75)

# - Plot these quartiles using vertical lines on a boxplot or histogram.

sns.boxplot(x=numeric_column)
plt.axvline(q25, color ='green')
plt.axvline(q50, color ='yellow')
plt.axvline(q75, color ='blue')
plt.title('Boxplot with Quartiles')
plt.xlabel('Total Bill')
plt.ylabel('Value')
plt.show()
```

```python
# Q.3
# - Print the five-number summary (min, Q1, median, Q3, max) of a column.
five_number_summary = {
    'Min': np.min(numeric_column),
    'Q1': q25,
    'Median': q50,
    'Q3': q75,
    'Max': np.max(numeric_column)
}
print(f"Five-Number Summary: {five_number_summary}")

# - Create a boxplot using seaborn.boxplot() or matplotlib.

sns.boxplot(x=numeric_column)
plt.title('Boxplot of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Value')
plt.show()

# - Identify and label outliers from the boxplot.
```

# - In a markdown cell, define inferential statistics and list real-world use cases (e.g., polling, clinical trials, A/B testing).
# - Load a sample population dataset.
# - Take a random sample using df.sample() and compare the sample mean to the population mean.
# - Repeat sampling multiple times and visualize the sample mean distribution.

Inferential Statistics is a branch of statistics that uses data from a sample to make generalizations, predictions, or decisions about a larger population. Instead of examining every individual in a population (which is often impractical), inferential statistics helps us draw conclusions and estimate population parameters based on sample data.

It involves:

- Estimating population parameters (e.g., mean, proportion)

-Hypothesis testing
 
- Making predictions or decisions using data

| **Domain**                 | **Use Case**           | **How Inferential Statistics Is Used**                                     |
| -------------------------- | ---------------------- | -------------------------------------------------------------------------- |
| 🗳️ **Polling**            | Election forecasts     | Estimating voter preferences from sample surveys                           |
| 🧪 **Clinical Trials**     | Drug effectiveness     | Determining if a new drug performs better than a placebo                   |
| 💻 **A/B Testing**         | Website optimization   | Comparing user responses to two versions of a web page                     |
| 📉 **Market Research**     | Product feedback       | Generalizing customer opinions from a sample to the broader market         |
| 🏥 **Public Health**       | Disease spread studies | Estimating infection rates and risk factors from sample data               |
| 🎓 **Education**           | Test score analysis    | Estimating average performance across schools or regions                   |
| 🛒 **Retail & E-commerce** | Sales forecasting      | Predicting future sales based on sample transaction data                   |
| 📊 **Manufacturing**       | Quality control        | Inferring the defect rate of an entire production batch from a few samples |

```python
population = np.random.normal(loc=50, scale=20, size=100)
print(population[:10])

population_mean = np.mean(population)
print(f"Population Mean: {population_mean}")

sample = np.random.normal(loc=50, scale=20, size=30)
print(sample[:10])

sample_mean = np.mean(sample, ddof=0)
print(f"Sample Mean: {sample_mean}")

plt.boxplot(population, vert=False)
plt.axvline(population_mean, color='red', label='Population Mean')
plt.axvline(sample_mean, color='blue', label='Sample Mean')
plt.title('Boxplot with Population and Sample Means')
plt.xlabel('Value')
plt.legend()
plt.show()
```
