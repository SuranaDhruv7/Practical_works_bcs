```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
# Q.1
# - Generate two lists of student scores for math and physics.

student_scores_math = np.random.randint(50, 100, size=30)
student_scores_physics = np.random.randint(50, 100, size=30)
print("Math Scores:", student_scores_math)
print("Physics Scores:", student_scores_physics)
print("\n-------------------------------------\n")

# - Convert to Pandas DataFrame.

df_scores = pd.DataFrame({
    'Math': student_scores_math,
    'Physics': student_scores_physics
})

# - Calculate the covariance using .cov().

covariance = df_scores.cov().iloc[0, 1]
print("Covariance between Math and Physics scores:", covariance)
print("\n-------------------------------------\n")

# - Calculate Pearson correlation coefficient using .corr(method='pearson').

pearson_corr = df_scores.corr(method='pearson').iloc[0, 1]
print("Pearson correlation coefficient between Math and Physics scores:", pearson_corr)
print("\n-------------------------------------\n")

# - Shuffle one of the score lists randomly.

df_scores['Physics_Shuffled'] = df_scores['Physics'].sample(frac=1).reset_index(drop=True)
print("Shuffled Physics Scores:\n", df_scores['Physics_Shuffled'])
print("\n-------------------------------------\n")

# - Now calculate Spearman rank correlation using .corr(method='spearman').

spearman_corr = df_scores.corr(method='spearman').iloc[0, 2]
print("Spearman rank correlation coefficient between Math and shuffled Physics scores:", spearman_corr)

# - Discuss the difference between Pearson and Spearman.

# Pearson correlation measures linear relationships, while Spearman correlation
# measures monotonic relationships. Pearson is sensitive to outliers, while Spearman
# is more robust to them. In this case, the Pearson correlation might be higher if
# the relationship is linear, while Spearman correlation will capture any monotonic
# relationship regardless of linearity. If the scores are shuffled, we expect both
# correlations to be close to zero, indicating no relationship.
```

```python
# Q.2
# - Generate and plot a uniform distribution using NumPy.

uniform_data = np.random.uniform(low=0, high=100, size=1000)
sns.histplot(uniform_data, bins=30, kde=True, color='blue', stat='density')
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# - Use a histogram with 50 bins.

sns.histplot(uniform_data, bins=50, kde=True, color='blue', stat='density')
plt.title('Uniform Distribution with 50 Bins')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# - Generate a binomial distribution using NumPy or scipy.stats.binom.

binomial_data = np.random.binomial(n=10, p=0.5, size=1000)
sns.histplot(binomial_data, bins=30, kde=False, color='green', stat='density')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Density')
plt.show()

# - Compare it visually with a normal distribution.

normal_data = np.random.normal(loc=5, scale=2, size=1000)
sns.histplot(normal_data, bins=30, kde=True, color='red', stat='density', label='Normal Distribution')
sns.histplot(binomial_data, bins=30, kde=False, color='green', stat='density', label='Binomial Distribution')
plt.title('Comparison of Normal and Binomial Distributions')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()

# - Simulate a Poisson distribution of daily customer counts.

poisson_data = np.random.poisson(lam=20, size=1000)
sns.histplot(poisson_data, bins=30, kde=False, color='purple', stat='density')
plt.title('Poisson Distribution (λ=20)')
plt.xlabel('Daily Customer Counts')
plt.ylabel('Density')
plt.show()

# - Plot and interpret.

sns.histplot(poisson_data, bins=30, kde=True, color='purple', stat='density')
plt.title('Poisson Distribution with KDE')
plt.xlabel('Daily Customer Counts')
plt.ylabel('Density')
plt.show()
```

```python
# Q.3
# - Generate a population with a skewed distribution (e.g., exponential).

population_data = np.random.exponential(scale=2, size=1000)
sns.histplot(population_data, bins=30, kde=True, color='orange', stat='density')
plt.title('Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# Draw 1000 samples of size 30 and calculate sample means.

sample_means = []
for i in range(1000):
    sample = np.random.choice(population_data, size=30, replace=True)
    sample_means.append(np.mean(sample))
sns.histplot(sample_means, bins=30, kde=True, color='blue', stat='density')
plt.title('Distribution of Sample Means (n=30)')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.show()

# - Plot the histogram of sample means and observe a normal shape.

sns.histplot(sample_means, bins=30, kde=True, color='blue', stat='density')
plt.title('Distribution of Sample Means (n=30) - Central Limit Theorem')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.axvline(np.mean(sample_means), color='red', linestyle='dashed', linewidth=1, label='Mean of Sample Means')
plt.legend()
plt.show()

# - Increase the sample size to 100 per sample.

sample_means_large = []
for i in range(1000):
    sample = np.random.choice(population_data, size=100, replace=True)
    sample_means_large.append(np.mean(sample))
sns.histplot(sample_means_large, bins=30, kde=True, color='green', stat='density')
plt.title('Distribution of Sample Means (n=100) - Central Limit Theorem')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.axvline(np.mean(sample_means_large), color='red', linestyle='dashed', linewidth=1, label='Mean of Sample Means')
plt.legend()
plt.show()

# - Show how CLT strengthens with larger sample sizes.

sns.histplot(sample_means_large, bins=30, kde=True, color='green', stat='density')
plt.title('Distribution of Sample Means (n=100) - Central Limit Theorem')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.axvline(np.mean(sample_means_large), color='red', linestyle='dashed', linewidth=1, label='Mean of Sample Means')
plt.legend()
plt.show()
```
